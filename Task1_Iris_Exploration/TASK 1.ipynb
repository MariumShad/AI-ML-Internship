# Task 1: Exploring and Visualizing a Simple Dataset

Objective:
Learn how to load, inspect, and visualize a dataset to understand data trends and distributions.

# Data Loading
We load the Iris dataset using seaborn.

```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df=sb.load_dataset("iris")
```

# Data Inspection
We explore dataset structure using shape, info and describe statistics.

```python
shape=df.shape
print("Shape of dataset: ",shape) #it will tell the col and row of dataset
```

```python
column_name=df.columns
print("Column Names: ",column_name) #it will tell the col names present in dataset
```

```python
first_few_rows=df.head()
print("First few rows: ",first_few_rows) #head gives first 5 values of dataset
```

```python
info=df.info()
print("Dataset Info: ",info) #info tells all details of dataset
```

```python
des=df.describe()
print("Description of dataset: ",des) #describe tells the description of dataset
```

# Data Visualization
We visualize the dataset using scatter plots, histograms and box plots to understand feature relationship, distributions and outliers.

```python
#SCATTER PLOT - it shows relationship between two features sepal_length and sepal_width
sb.scatterplot(x="sepal_length",y="sepal_width",hue="species",data=df)
plt.title("SCATTER PLOT")
plt.show()
```

```python
#HISTOGRAM - it creates distribution plots for each numerical feature
df.hist(figsize=(10,8))
plt.suptitle("HISTOGRAM")
plt.show()
```

```python
#BOXPLOT - it shows outliers
sb.boxplot(data=df)
plt.title("BOX PLOT FOR OUTLIERS")
plt.show()
```
